```python
import requests
from bs4 import BeautifulSoup

# CSS Selector 자동 생성 함수
def get_unique_selector(element):
    path = []
    while element.parent:
        siblings = element.parent.find_all(element.name, recursive=False)
        if len(siblings) > 1:
            index = siblings.index(element) + 1
            path.append(f'{element.name}:nth-of-type({index})')
        else:
            path.append(element.name)
        element = element.parent
    return ' > '.join(reversed(path))

url = 'https://www.moneys.co.kr/article/2024052410400231613'

response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.text, 'lxml')

# 본문 중 일부 문장을 입력하면 됩니다.
target_text_snippet = '법인차는 말 그대로 개인이 아닌 법인이 구매, 사용하는 차를 의미한다.'

target_element = soup.find(lambda tag: tag.name and tag.get_text() and target_text_snippet in tag.get_text())

if target_element:
    selector = get_unique_selector(target_element)
    print("자동으로 찾은 CSS Selector:", selector)
    
    # 자동 생성한 selector로 본문 크롤링
    content_tag = soup.select_one(selector)
    content = content_tag.get_text(separator='\n', strip=True)
    
    print("\n본문 내용:\n", content)
else:
    print("본문 요소를 찾지 못했습니다.")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

# 크롬 브라우저 옵션 설정 (백그라운드로 실행)
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# 웹 드라이버 생성
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 크롤링할 구글 뉴스 URL
url = 'https://news.google.com/search?q=연두색번호판&hl=ko&gl=KR&ceid=KR%3Ako'

driver.get(url)
time.sleep(3)  # 페이지 로딩을 위한 시간 확보

# 뉴스기사 전체 크롤링
articles = driver.find_elements(By.CSS_SELECTOR, 'article')

news_data = []

for article in articles[:5]:  # 앞 5개 기사만 예시로 가져옴
    try:
        # 기사 제목
        title_el = article.find_element(By.CSS_SELECTOR, 'h3, h4')
        title = title_el.text

        # 기사 클릭 후 본문 추출
        driver.execute_script("arguments[0].click();", title_el)
        time.sleep(2)
        
        # 새로 열린 탭으로 전환
        driver.switch_to.window(driver.window_handles[-1])
        time.sleep(2)

        # 열린 기사 본문 크롤링 (가장 일반적인 기사 본문 선택자)
        paragraphs = driver.find_elements(By.CSS_SELECTOR, 'p')

        content = "\n".join(p.text for p in paragraphs if p.text.strip())

        news_data.append({'title': title, 'content': content})

        driver.close()
        driver.switch_to.window(driver.window_handles[0])

    except Exception as e:
        print(f"에러가 발생했습니다: {e}")
        continue

driver.quit()

# 크롤링 결과 출력
for idx, news in enumerate(news_data, 1):
    print(f'\n{"="*50}\n📌 [{idx}] {news["title"]}\n{"="*50}')
    print(news["content"][:1000], '...')  # 내용이 길 경우 1000자까지만 출력
```
